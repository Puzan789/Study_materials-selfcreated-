```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Draw a line in a diagram from position (0,0) to position (6,250)
x=np.array([0,6])
y=np.array([0,250])
plt.plot(x,y)
plt.show()
```

```python
# The plot() function is used to draw points (markers) in a diagram.

# By default, the plot() function draws a line from point to point.

# The function takes parameters for specifying points in the diagram.

# Parameter 1 is an array containing the points on the x-axis.

# Parameter 2 is an array containing the points on the y-axis.

# If we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.
```

```python
# Draw a line in a diagram from position (1,3) to (8,10)
```

```python
x=np.array([1,8])
y=np.array([3,10])
plt.plot(x,y)
plt.show()
```

### Plotting Without a Line


```python
# To plot only the markers you can use shortcut string notation parameter 'o' which means 'rings'
```

```python
x=np.array([1,8])
y=np.array([3,10])
plt.plot(x,y,'o')
plt.show()
```

## Multiple Points

```python
# You can plot as many points as you like , just make sure you have the same numbers of point in both the axis
```

```python
#Draw a line in a Diagram from position (1,3) to (2,8) then to (6,1) and finally to position (8,10)
```

```python
x=np.array([1,2,6,8])
y=np.array([3,8,1,10])
plt.plot(x,y,)
plt.show()
```

## Default x-Points 

```python
# if we donot specify the points on the x-axis ,
#they will get the default value 0,1,2,3,(etc., depending on the length of the y -points)
```

```python
y=np.array([4,2,4,2])
plt.plot(y)
plt.show()
```

## Matplotlib Markers

```python
# you can use the keyword argument marker to emphasize each point with a specified marker
```

```python
y=np.array([2,5,6,8])
plt.plot(y,marker="d")
plt.show()
```

## Format Strings (fmt) 

```python
# you can also use the shortcut string notation parameter to specify the marker.
# This parameter is also called fmt and is written with this syntax 'marker|line|color'
```

```python
y=np.array([3,2,5])
plt.plot(y,'o-.k')
plt.show()
```

## Marker Size

```python
# You can use the keyword markersize or the shorter version ms to set the size of the markers
```

```python
# Set the size of markers to 20
```

```python
y=np.array([6,7,4,6])
plt.plot(y,'o-.k',ms=20)
plt.show()
```

## markeredgecolort and markerfacecolor

```python
# You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers
```

```python
y=np.array([3,2,5])
plt.plot(y,'o-.',mec="r",mfc="g",ms=20)
plt.show()
```

## Matplotlib Line

###  you can use the keyword linestyle,or shorter ls to change the style of the plotted line:
### you can use the keyword argument color or the shorter c to set the color of the line.
### you can use the keyword argument linewidth or the shorter lw to change the width of the line.


```python

x=np.array([1,2,3,5])
y=np.array([4,5,2,3])
plt.plot(x,y,ls="solid",color="#e52195",lw=10)
plt.show()
```

# Multiple Lines

#### You can plot as many lines as you like by simply adding more plt.plot()

```python
y1=np.array([9,2,3,4,5])
y2=np.array([5,3,4,2,1])
plt.plot(y1,color="red")
plt.plot(y2)
plt.show()
```

# Matplotlib Labels and Titles

```python
x=np.array([12,3,43,12,32,43,54,2])
y=np.array([34,54,23,54,23,45,65,76])
plt.plot(x,y,'o:',mfc="g",mec="red",color="#ff00ff",lw="10")
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x,y)
plt.xlabel('going')
plt.ylabel('coming')
plt.title("Just a path")
plt.show()
```

### Set font property ,Location

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x,y)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.xlabel('going',fontdict=font1)
plt.ylabel('coming',fontdict=font2)
plt.title("Just a path",fontdict=font1,loc='right')
plt.show()
```

###  Add Grid

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x,y)
plt.xlabel('going')
plt.ylabel('coming')
plt.title("Just a path")
plt.grid(axis='x',color="g",linestyle="dotted",lw="4")
plt.show()
```

## Subplot

### With a subplot() function you can draw multiple plots in one figure

```python
#plot1
x=np.array([0,1,2,3])
y=np.array([2,3,1,4])
plt.subplot(1,3,1)
plt.plot(x,y)
x=np.array([0,12,2,3])
y=np.array([2,33,1,4])
plt.subplot(1,3,2)
plt.plot(x,y)
x=np.array([0,12,2,3])
y=np.array([2,33,1,4])
plt.subplot(1,3,3)
plt.plot(x,y)
# You can add a title for all figure as super title
plt.suptitle("randoms")
```
