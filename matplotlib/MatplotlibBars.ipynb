```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array(["a","b","c"])
y=np.array([2,3,4])
plt.bar(x,y)
plt.show()
```

```python
# If you want to plot bar horizontally instead of vertically you can use barh
```

```python
x=np.array(["a","b","c"])
y=np.array([2,3,4])
plt.subplot(1,2,1)
plt.barh(x,y)
plt.show()
```

## Bar Color

```python
x=np.array(["A","B","c"])
y=np.array([3,2,4])
plt.bar(x,y,color="#4C5169")
plt.title("Bars")
plt.show()
```

### BarWidth

```python
x=np.array(["A","B","c"])
y=np.array([3,2,4])
plt.bar(x,y,color="#cad32f",width=0.4)
plt.title("Bars",loc="Left")

plt.show()
```

### For horizontal bars use height instead of width

```python
x=np.array(["A","B","c"])
y=np.array([3,2,4])
plt.barh(x,y,color="#4C5169",height=0.2)
plt.title("Bars")
plt.show()
```
