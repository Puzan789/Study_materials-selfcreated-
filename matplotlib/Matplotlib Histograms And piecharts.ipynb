# PieCharts

```python
# With pyplot , you can use the pie() functions to draw piecharts
```

```python
y=np.array([23,4,21,34])
plt.pie(y,labels=['a','b','c','d'])
plt.show()
```

```python
# By default startangle is 0 but you can change to 90 180 270 
```

```python
y=np.array([23,4,21,34])
plt.pie(y,labels=['a','b','c','d'],startangle=90)
plt.show()
```

## Explode
### Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.


```python
x=np.array(['apple','bananas','Cats'])
y=np.array([1,3,4])
myexplode = [0.2, 0.5, 0.9]
colors=['#430302','#222312',"#326523"]
plt.pie(y,labels=x,explode=myexplode,shadow=True,colors=colors)
plt.legend(title="Anything")
plt.show()
```
