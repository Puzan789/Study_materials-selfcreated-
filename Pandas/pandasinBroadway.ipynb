```python
import csv
```

```python
mydict=[{
    'name':['ram','shyam','hari'],
    'class':[3,4,5],
    "address":["ilam",'jhapa','brt']
}]
fields=['name','class','address']
file=open('data.csv','w')
writer=csv.DictWriter(file,fieldnames=fields)
writer.writeheader()
for i in mydict:
    writer.writerow(i)
```
